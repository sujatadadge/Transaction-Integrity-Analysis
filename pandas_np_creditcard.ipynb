```python
#drop duplicates
df.drop_duplicates(inplace=True)
```

```python
print(df)
```

cleaned Data - no null values , duplicates removed

```python
#Dataset summary & class distribution
#fraud vs Non-fraud
class_counts = df['Class'].value_counts()
print(class_counts)
```

```python
fraud_percent = (class_counts[1] / len(df)) * 100
print(f"Fraud Percentage: {fraud_percent:4f}%")
```

Observation:
Only 0.17% transactions are fraudulent â€” this shows massive class imbalance, which is a real-world financial challenge.

```python
#Amount Analysis
#seprated Data
fraud = df[df['Class'] == 1]
non_fraud = df[df['Class'] == 0]

#Median Comparison
print("Average Fraud Amount:", np.mean(fraud['Amount']))
print("Average Non-fraud Amount:", np.mean(non_fraud['Amount']))
```

```python
#Median comparison
print("Median Fraud Amount:", np.median(fraud['Amount']))
print("Median Non-Fraud Amount:", np.median(non_fraud['Amount']))
```

Observation:
Fraud transactions tend to have smaller median amounts but show high variance, meaning some frauds are large-value spikes.

```python
print(df['Time'].describe())
```

```python
#Average time differnce between fraud and non fraud
print("fraud Time Mean:", np.mean(fraud['Time']))
print("non_fraud Mean:", np.mean(non_fraud['Time']))
```

```python
#checkimg distrubtion by time intervals
df['Hour'] = (df['Time']/ 3600 ).astype(int)
hourly_fraud = df.groupby('Hour')['Class'].sum()
hourly_fraud.head(24)
```

Observation:
Fraud transactions may cluster during specific hours â€” useful for fraud detection teams to monitor risky time slots.

```python
#Correlation Study(correlation of all features)
corr_matrix = df.corr()
```

```python
#Top correlations with class
corr_target = corr_matrix['Class'].sort_values(ascending = False)
corr_target.head(10)
```

Observation:
Some PCA components (e.g., V17, V14, V12) show strong correlation with fraud â€” these could be important features for predictive models.

```python
#statistical Exploration
#Mean , median and std for all numeric column
stats = df.describe().T[['mean','std','min','max']]
stats['range'] = df.isnull().sum()
stats.head()
```

Observation:
Statistical summary shows that feature ranges vary widely, justifying PCA normalization before modeling.

```python
#variance and range using numpy
variance = np.var(df['Amount'])
rng = np.ptp(df['Amount']) #Peakto-peak range
```

```python
print("Variance of Amount:", variance)
```

```python
print("Range of Amount:",rng)
```

```python
#Manual standard deviation check
manual_std = np.sqrt(np.mean((df['Amount'] - np.mean(df['Amount']))** 2))
print("Manual std Dev(check):", manual_std)
```

Observation:
Variance and range are extremely high â€” confirming skewed distribution in transaction amounts.

```python
#Group & Aggragation Analysis
#group transactions by hours and calculate mean amount
hourly_amount = df.groupby('Hour')['Amount'].mean().round(2)
```

```python
#fraud rate per hour
fraud_rate_per_hour = df.groupby('Hour')['Class'].mean().round(4)
```

```python
print("hourly Avg Amount:\n", hourly_amount.head(10))
```

```python
print("\nFraud rate per Hour:\n", fraud_rate_per_hour.head(10))
```

Observation:
Fraud frequency might rise during low-activity hours (night/midnight range).

```python
summary = {
    "Total Transactions": len(df),
    "Fraud Transactions": len(fraud),
    "Non-Fraud Transactions": len(non_fraud),
    "Fraud Percentage": round(fraud_percent, 3),
    "Avg Fraud Amount": round(np.mean(fraud['Amount']), 2),
    "Avg Non-Fraud Amount": round(np.mean(non_fraud['Amount']), 2),
    "Highest Fraud Amount": np.max(fraud['Amount'])
}

for k, v in summary.items():
    print(f"{k}: {v}")
```
